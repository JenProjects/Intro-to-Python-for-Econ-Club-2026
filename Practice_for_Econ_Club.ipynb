```python
#Starting off with numpy
import numpy as np
#an array is a collection of elements, typically of the same data type, stored at contiguous memory locations
#contiguous means next together in sequence
gdp_growth = np.array([2.1, 2.5, 1.8, 2.9, 3.2])
print(gdp_growth)

print("Average GDP growth:", np.mean(gdp_growth))
print("Maximum GDP growth:", np.max(gdp_growth))
print("Minimum GDP growth:", np.min(gdp_growth))
print("GDP growth plus 1:", gdp_growth + 1)
```

NumPy helps us work with numerical data efficiently. Instead of typing calculations one by one, we can store values in an array and quickly calculate things like the mean, max, or min.

```python
#Next is Pandas
import pandas as pd

data = { "Year": [2019, 2020, 2021, 2022, 2023],
    "GDP_Growth": [2.1, -3.4, 5.9, 2.1, 2.5],
    "Unemployment": [3.7, 8.1, 5.4, 3.6, 3.5],
    "Inflation": [1.8, 1.2, 4.7, 8.0, 4.1]}

df = pd.DataFrame(data)
print(df)
#This is a Python dictionary where each key represents a variable and each value is a list of observations
#We can easily convert into a dataset using pandas.
```

```python
#if time permits go through counting in python
data["GDP_Growth"][1]
```

```python
#do these one by one
print(df.head(1))
print(df.describe())
print(df["Unemployment"])
```

```python
#appends it to the end of the table
df["Real_Stress"] = df["Inflation"] + df["Unemployment"]
print(df)
```

Pandas is probably the most useful library for many economics students because it works with data tables. If you’re used to Excel, this is the closest Python equivalent. We can create a DataFrame, inspect the data, summarize it, and even create new variables.

```python
#only if have time
#sort_values() organizes the dataset by a chosen column, and ascending=False tells Python to display the highest values first.
print(df.sort_values("Inflation", ascending=False))
```

```python
#Switching to Matplotlib
import matplotlib.pyplot as plt

plt.plot(df["Year"], df["Unemployment"])
plt.title("Unemployment Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate")
```

This is where Python becomes really helpful for research presentations and papers. Once the data is organized, we can easily create graphs to spot trends and communicate findings.

```python
#Maybe show alternative bar chart if have time
plt.bar(df["Year"], df["Inflation"])
plt.title("Inflation by Year")
plt.xlabel("Year")
plt.ylabel("Inflation Rate")
plt.show()
```

```python
#Last library is seaborn
import seaborn as sns
titanic = sns.load_dataset("titanic")

#go through one by one and ask students what they can tell by looking at the output
#ask about
print(titanic.columns)
print(titanic.info())
print(titanic.describe())
```

Introducing the question: “Did passenger class affect survival?” We want to see average survival rate for each passenger class.

```python
#Ask class what they can infer from the output
titanic.groupby("class")["survived"].mean()
```

```python
#palette makes things prettier
sns.barplot(x="class", y="survived", data=titanic, palette='rainbow')

#errorbar=None removes the black error line
plt.title("Survival Rate by Passenger Class")
```

```python
#Divided into 20 bins (rectangles)
sns.histplot(titanic["age"], bins=20)

plt.title("Distribution of Passenger Ages")
```

```python
sns.boxplot(x="class", y="age", data=titanic)

plt.title("Age Distribution by Passenger Class")
```

```python
#Splitting by sex
sns.catplot(x="class", y="survived", hue="sex", kind="bar", data=titanic, palette ='coolwarm')
plt.title("Survival Rate by Passenger Class and Gender")
plt.show()
```

A catplot, or categorical plot, is used to compare groups in a dataset. Here we’re comparing survival rates across passenger classes. Because survival is coded as 0 or 1, the average value represents the probability of survival. We can see that first-class passengers had the highest survival rate.

```python
#violin plot
sns.violinplot(x="class", y="age", data=titanic)

plt.title("Age Distribution by Passenger Class")
```

A violin plot shows the distribution of a variable across categories. In this case, we’re looking at the age distribution of passengers in each class. The width of the shape shows where more observations occur. Wider sections indicate that more passengers fall into that age range.

```python
corr = titanic.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix of Titanic Variables")
plt.show()
```

A heatmap shows the correlations between numerical variables in a dataset. The color intensity represents the strength of the relationship. Red indicates a positive relationship and blue indicates a negative one. This is a quick way researchers can identify patterns in data.

```python
#First do without hue and next do with hue
sns.lmplot(x="age", y="fare",  hue="class", data=titanic)

plt.title("Age vs Fare by Passenger Class")
```
